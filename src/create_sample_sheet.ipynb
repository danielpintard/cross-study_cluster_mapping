```python
import pandas as pd
import numpy as np
import os

res_path = "/data/pintardde/class_marker_pipeline/results/"
```

```python
studies = []
for folder in os.listdir(res_path):
    if not any(x in folder for x in ['ds_', 'downsample', 'normalized', 'split_by_study']):
        studies.append(folder)
```

```python
study_x_sample_info = {
    study: []
    for study in studies
}

base_h5ad_path = "/data/pintardde/class_marker_pipeline/data"
base_nsf_res_path = "/data/pintardde/class_marker_pipeline/results/"

for k, v in study_x_sample_info.items():
    study_res_id = k
    study_id = study_res_id.rsplit("_", 1)[0]
    # add path to h5ad file
    if k == 'LungMap_CellRef_results': v.append(os.path.join(base_h5ad_path, f"{study_id}(edited_var).h5ad"))
    else: v.append(os.path.join(base_h5ad_path, f"{study_id}.h5ad"))
    # add path to NS-Forest results
    v.append(os.path.join(base_nsf_res_path, study_res_id))
    
for k, v in study_x_sample_info.items():
    # give cxg flag
    if k in ['33208946_travaglini_2020_nature_human_10x_results', 
             '33208946_travaglini_2020_nature_human_smartseq2_results', 
             '39818203_niethamer_2025_cellstemcell_mouse_results',
             'HLCA_Core_results', 'tabula_senis_mouse_10x_results', 'tabula_senis_mouse_smartseq2_results']:
        v.append(True)
    else:
        v.append(False)

for k, v in study_x_sample_info.items():
    study_id = k.rsplit("_", 1)[0]
    # add .var params
    if study_id == 'LungMap_CellRef': v.append("features")
    else: v.append("")
    
    # add cluster_header param
    if study_id == 'LungMap_CellRef': v.append("celltype_level3_fullname")
    elif study_id == 'HLCA_Core': v.append("ann_finest_level")
    elif study_id == "39818203_niethamer_2025_cellstemcell_mouse": v.append("author_cell_type") 
    
    else: v.append("free_annotation")
    
```

```python
sheet_cols = ["h5ad_path", "markers_path", "cxg_flag", "var_col_arg", "cluster_header_arg"]
sample_sheet = pd.DataFrame(study_x_sample_info.values(), index=study_x_sample_info.keys(), columns = sheet_cols)
sample_sheet.index.rename("data_id", inplace=True)
```
